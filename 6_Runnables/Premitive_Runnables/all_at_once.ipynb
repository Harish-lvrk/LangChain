## RunnableLambda

Alright, let‚Äôs dive into **RunnableLambda** step by step, like proper class notes.

---

### 1. Introduction / Definition

**RunnableLambda** is a special Runnable in LangChain that lets you **wrap any normal Python function** (lambda or def) into the Runnable interface.

That means:

* Your own custom code can now behave exactly like an LLM, a PromptTemplate, or a Retriever.
* It supports `.invoke()`, `.batch()`, `.stream()`, and can be composed with `|` just like other Runnables.

---

### 2. Why it is Important

‚Ä¢ In pipelines, not everything is an LLM or retriever. Sometimes you just need small steps like:

* cleaning text
* formatting data
* adding numbers
* custom business logic

RunnableLambda gives a **bridge**: your Python code becomes a first-class Runnable, so you can insert it seamlessly into chains.

Without it ‚Üí you‚Äôd need ‚Äúglue code‚Äù outside LangChain.
With it ‚Üí everything fits into the same Runnable system.

---

### 3. How it Works (intuitive explanation)

Imagine you have a **conveyor belt of tasks** (runnables).

* Normally, each station is a special machine (LLM, retriever, parser).
* But sometimes, you just need a human with scissors to trim something simple.

RunnableLambda = *that human station*.
It takes an input, applies your Python function, and passes the result forward.

---

### 4. Code Examples

A) Simple Math Example

```python
from langchain_core.runnables import RunnableLambda

# Wrap a Python function
add_one = RunnableLambda(lambda x: x + 1)

print(add_one.invoke(5))  # Output: 6
```

B) Composition with |

```python
double = RunnableLambda(lambda x: x * 2)
pipeline = add_one | double   # Equivalent to double(add_one(x))

print(pipeline.invoke(3))  # (3+1)*2 = 8
```

C) Batch Processing

```python
print(add_one.batch([1, 2, 3]))  
# Output: [2, 3, 4]
```

D) Text Preprocessing in a Chain

```python
from langchain_core.prompts import PromptTemplate

# Function: lowercase text
lowercase = RunnableLambda(lambda s: s.lower())

# Combine with prompt
prompt = PromptTemplate.from_template("Summarize: {text}")

chain = lowercase | prompt
print(chain.invoke("HELLO WORLD"))  
# "Summarize: hello world"
```

---

### 5. Real-life Analogy

Think of RunnableLambda like a **‚Äúcustom adapter‚Äù in a machine line**.

* You can take any small helper (calculator, text cleaner, reformatter)
* Plug it in so it behaves like any other official machine in the factory.

It‚Äôs like hiring a freelancer who agrees to follow the company‚Äôs standard process ‚Äî so you don‚Äôt have to treat them differently.

---

### 6. Applications / Use Cases

* Preprocessing (e.g., lowercase, strip punctuation, JSON cleaning)
* Postprocessing (e.g., format LLM output into custom structure)
* Data enrichment (e.g., add extra metadata before calling LLM)
* Routing logic (e.g., decide which branch to send input to)
* Quick testing (try ideas without writing a full custom Runnable class)

---

### 7. Advantages and Limitations

**Advantages:**

* Very flexible ‚Äî any Python function can be integrated.
* No need to subclass or write complex wrappers.
* Plays nicely with batching, streaming, and composition.

**Limitations:**

* If the function is too heavy/slow, it doesn‚Äôt add scaling benefits by itself.
* Unlike specialized Runnables (LLM, Retriever), you must ensure your function handles inputs/outputs correctly.

---

### 8. Final Summary

RunnableLambda = **the glue between LangChain and your custom Python code**.

* Wraps any function into the Runnable system.
* Supports invoke, batch, stream, and composition.
* Lets you seamlessly mix your own logic with LLMs, retrievers, prompts, and parsers.

Think of it as your **‚Äúcustom mini-tool‚Äù** inside a larger AI workflow.

## RunnableBranch
Alright, let‚Äôs unpack **Conditional Runnables** in LangChain like a proper class lecture.

---

### 1. Introduction / Definition

Conditional Runnables let you **choose different execution paths** in a pipeline depending on the input.

Instead of always running the same sequence, you can add an **if/else-like logic**:

* If condition A ‚Üí run Runnable X
* Else if condition B ‚Üí run Runnable Y
* Otherwise ‚Üí run Runnable Z

In LangChain, the main tool for this is **RunnableBranch**.

---

### 2. Why It Is Important

In real-world workflows, not every input should be handled the same way:

‚Ä¢ Example 1: In customer support

* If the query is about billing ‚Üí send to Billing LLM.
* If it‚Äôs about technical issues ‚Üí send to Tech LLM.

‚Ä¢ Example 2: In RAG (retrieval-augmented generation)

* If the question is answerable from docs ‚Üí run Retriever ‚Üí LLM.
* If not ‚Üí fallback to a general knowledge LLM.

‚Ä¢ Example 3: In moderation pipelines

* If input is toxic ‚Üí block or re-route.
* If input is safe ‚Üí continue with normal processing.

Without conditional runnables, you‚Äôd have to manually code all this branching logic outside the chain.

---

### 3. How It Works (Intuitive Explanation)

Think of it like a **traffic cop at an intersection**:

* Each car (input) comes in.
* The cop checks the condition (is the car small, big, emergency?).
* Depending on the rule, the cop sends it down the left road, the right road, or keeps it straight.

RunnableBranch is that traffic cop inside LangChain.

---

### 4. Types / Variants

Currently, LangChain provides mainly:

‚Ä¢ **RunnableBranch** ‚Äì define a list of `(condition, runnable)` pairs + a default runnable.

* First condition that matches decides the branch.
* If no condition matches, the default branch runs.

‚Ä¢ **Custom Conditional Logic** ‚Äì sometimes built with `RunnableLambda` to preprocess inputs, then feed them into a branch.

---

### 5. Code Examples

A) Simple Even/Odd Example

```python
from langchain_core.runnables import RunnableLambda, RunnableBranch

# Define conditions
is_even = lambda x: x % 2 == 0
even_chain = RunnableLambda(lambda x: f"{x} is even")
odd_chain = RunnableLambda(lambda x: f"{x} is odd")

# Create branch
branch = RunnableBranch(
    (is_even, even_chain),
    RunnableLambda(lambda x: f"{x} is odd (default)")  # default
)

print(branch.invoke(4))  # "4 is even"
print(branch.invoke(7))  # "7 is odd (default)"
```

B) Realistic: Route Customer Queries

```python
from langchain_core.runnables import RunnableLambda, RunnableBranch

def is_billing(query):
    return "bill" in query.lower()

def is_tech(query):
    return "error" in query.lower() or "issue" in query.lower()

billing_llm = RunnableLambda(lambda q: f"Billing response for: {q}")
tech_llm = RunnableLambda(lambda q: f"Tech support response for: {q}")
default_llm = RunnableLambda(lambda q: f"General response for: {q}")

router = RunnableBranch(
    (is_billing, billing_llm),
    (is_tech, tech_llm),
    default_llm
)

print(router.invoke("I have an error with my account"))
# Output: Tech support response for: I have an error with my account
```

---

### 6. Real-life Analogy

‚Ä¢ Traffic Cop ‚Üí routes cars to different roads based on rules.
‚Ä¢ Hospital Receptionist ‚Üí patient comes in, receptionist checks the condition:

* Emergency ‚Üí ICU
* Child ‚Üí Pediatrics
* General ‚Üí OPD

RunnableBranch = receptionist for your data pipeline.

---

### 7. Applications / Use Cases

* **Routing Queries**: Different LLMs for different domains (billing, tech, general).
* **Moderation Pipelines**: Filter toxic vs non-toxic queries.
* **Fallback Mechanisms**: If retrieval finds docs, use them; otherwise, fallback to general LLM.
* **Multi-step workflows**: Choose next step dynamically (e.g., classification ‚Üí summarization vs translation).

---

### 8. Advantages and Limitations

**Advantages:**

* Flexible routing without leaving LangChain.
* Keeps pipelines modular and readable.
* Works seamlessly with other runnables (sequence, parallel, etc.).

**Limitations:**

* Condition functions must be written carefully (wrong condition = wrong path).
* Only first matching condition is used (no multiple matches).
* Complex branching trees can get messy to debug.

---

### 9. Final Summary

* **Conditional Runnables** allow if/else style branching inside LangChain pipelines.
* Main tool: **RunnableBranch** = list of `(condition, runnable)` pairs + default.
* Intuition: like a **traffic cop** deciding which road each input should take.
* Great for routing, moderation, fallbacks, and multi-domain workflows.
* Advantages: flexible and clean; Limitations: conditions must be carefully designed.

## LCEL

Alright Stark, let‚Äôs go step by step and build a **clear, intuitive explanation of LCEL (LangChain Expression Language)**.

---

## 1. Introduction / Definition

**LCEL (LangChain Expression Language)** is a **declarative way to build chains/pipelines** in LangChain using a simple syntax (like pipes `|`, dicts `{}`, and functions).

Instead of writing long custom classes or glue code, you express workflows **directly in code like math expressions**.

Example intuition:

* Math: `y = (x + 1) * 2`
* LCEL: `result = retriever | prompt | llm | parser`

So LCEL is **not a separate language** ‚Äì it‚Äôs just a clean, expressive API built on top of Python for composing **Runnables**.

---

## 2. Why It Is Important

Before LCEL:

* You had to use **Chains** (like `LLMChain`, `SimpleSequentialChain`, etc.).
* Each chain type had its own constructor, and mixing them was messy.
* Adding parallel, conditional, or batch processing required extra boilerplate.

With LCEL:

* Every component is a Runnable.
* You can connect them easily with pipes, dicts, or branches.
* It‚Äôs **declarative** ‚Üí you describe what you want, not how to glue it together.

---

## 3. How It Works (Intuitive Explanation)

Think of LCEL as **Lego blocks**:

* Each block = a Runnable (LLM, Prompt, Parser, Retriever, Lambda, etc.).
* The `|` operator = ‚Äúsnap blocks in a line‚Äù (sequence).
* A dict `{ "a": block1, "b": block2 }` = ‚Äúrun blocks in parallel‚Äù and collect results.
* A branch = ‚Äúif/else logic‚Äù for choosing paths.

So LCEL lets you **draw pipelines directly in code**.

---

## 4. Types of Composition in LCEL

1. **Sequential (| operator)**

   * Connects components in a straight line.
   * Example: `retriever | prompt | llm | parser`

2. **Parallel (dict)**

   * Run multiple components at once on the same input.
   * Example: `{ "summary": summarizer, "keywords": keyword_extractor }`

3. **Branching (RunnableBranch)**

   * Route inputs conditionally.
   * Example: `RunnableBranch([(is_question, qa_chain)], default=chat_chain)`

---

## 5. Code Examples

A) Simple Sequential Chain

```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

prompt = PromptTemplate.from_template("Tell me a joke about {topic}")
llm = ChatOpenAI(model="gpt-3.5-turbo")
parser = StrOutputParser()

chain = prompt | llm | parser

print(chain.invoke({"topic": "cats"}))
```

B) Parallel Example

```python
from langchain_core.runnables import RunnableParallel, RunnableLambda

chain = RunnableParallel({
    "original": RunnableLambda(lambda x: x),
    "length": RunnableLambda(lambda x: len(x)),
})

print(chain.invoke("LangChain is cool"))
# Output: {"original": "LangChain is cool", "length": 17}
```

C) Branch Example

```python
from langchain_core.runnables import RunnableBranch, RunnableLambda

is_question = lambda x: x.endswith("?")
qa = RunnableLambda(lambda x: f"Answering the question: {x}")
chat = RunnableLambda(lambda x: f"Chatting: {x}")

branch = RunnableBranch(
    (is_question, qa),
    chat  # default
)

print(branch.invoke("How are you?"))   # Answering the question: How are you?
print(branch.invoke("Nice weather today"))  # Chatting: Nice weather today
```

---

## 6. Real-life Analogy

Think of LCEL like **drawing a flowchart with code**:

* `|` = arrows connecting boxes.
* Dict `{}` = splitting into multiple branches that run in parallel.
* Branch = decision diamond (‚Äúyes ‚Üí this path, no ‚Üí that path‚Äù).

So instead of coding the wiring, you just **sketch your workflow in expressions**.

---

## 7. Applications / Use Cases

* **RAG Pipelines**: `retriever | prompt | llm | parser`
* **Multi-output generation**: `{ "summary": summarizer, "keywords": keyword_extractor }`
* **Conditional Routing**: send input to different LLMs or retrievers based on conditions.
* **Preprocessing + Postprocessing**: `lambda_cleaner | llm | parser`
* **Batch/Streaming friendly workflows**: everything supports `.batch()` and `.stream()`.

---

## 8. Advantages and Limitations

**Advantages:**

* Declarative and intuitive syntax.
* Works with all Runnables uniformly.
* Supports sync, async, batch, and streaming without extra code.
* Easy to visualize pipelines.

**Limitations:**

* Debugging large, complex graphs can still be tricky.
* Requires you to understand input/output shapes carefully (dict vs str vs list).
* Some advanced features (like retry/fallbacks) need extra wrappers.

---

## 9. Final Summary

* **LCEL = LangChain Expression Language**, a **declarative syntax for building pipelines**.
* Uses `|` for sequences, dicts `{}` for parallel, and branches for conditionals.
* Intuition: Lego blocks (Runnables) + arrows (`|`) = flowchart in code.
* Benefits: cleaner, composable, production-ready pipelines with less boilerplate.
* Real power: you can express **complex workflows** (retrieval, routing, enrichment, parsing) with just a few lines of code.
